```python
def boxplot(data, col1 = 'University of Hull', col2 = 'University of Cumbria',  fields = ['Clubs and Societies', 'Job Prospects', 'Student Support', 'college']):
    #list_of_fields = ['Accommodation', 'City Life', 'Clubs and Societies', 'Course and Lecturers', \
                      #'Facilities', 'Job Prospects', 'OVERALL UNIVERSITY RATING', 'Student Support', \
                      #"Students' Union",'Uni Facilities','college']
    
    vio = data[(data['college'] == col1) | (data['college'] == col2)][fields]
    
    
    x = vio['college']

    trace0 = go.Box(
        y=vio[fields[0]],
        x=x,
        name='Accommodation',
        marker=dict(
            color='#3D9970'
        )
    )
    trace1 = go.Box(
        y= vio[fields[1]],
        x=x,
        name='City Life',
        marker=dict(
            color='lightblue'
        )
    )
    trace2 = go.Box(
        y=vio[fields[2]],
        x=x,
        name='Clubs and Societies',
        marker=dict(
            color='#FF851B'
        )
    )
    data = [trace0, trace1, trace2]
    layout = go.Layout(
        yaxis=dict(
            title='normalized moisture',
            zeroline=False
        ),
        boxmode='group'
    )
    
    fig = go.Figure(data=data, layout=layout)
    plot(fig)
    return plot(fig, filename='static/temp-plot.html')
```
